$1$. Построить двумерный график на основе следующих уравнений:

$
\begin{align}
x = 16sin^3(t),
\end{align}
$

$
\begin{align}
y = 13cos(t) - 5cos(2t) - 2cos(3t) - cos(4t),
\end{align}
$

при $0 \leq t \leq 2\pi$.

При построении графика задать параметры:

*   `figsize=(6,6)`;
*   `color='red'`.

Дополнительно построить такой же график с внутренней цветовой заливкой при помощи метода `.fill()`.




```python
import math
import matplotlib.pyplot as plt


def x(t):
    return 16 * (math.sin(t) ** 3)


def y(t):
    return 13 * math.cos(t) - 5 * math.cos(2 * t) - 2 * math.cos(3 * t) - math.cos(4 * t)


gx = []
gy = []

t = 0
while t <= 2 * 3.14:
    gx.append(x(t))
    gy.append(y(t))
    t += 0.01

plt.figure(0, figsize=(6, 6))

plt.plot(gx, gy, color='red')
plt.fill(gx, gy, color='r')

plt.show()
```

$2$. Построить трёхмерный график по уравнению:

$
\begin{align}
(x^2 + \frac{9y^2}{4} + z^2 - 1)^3 - x^2z^3 - \frac{9y^2z^3}{200} = 0
\end{align}
$

При построении графика задать параметры:

*   `figsize=(10,10)`;
*   `colors=('red')`.

Фрагмент алгоритма отрисовки представлен ниже.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


#напишите код здесь
def func(x, y, z):
    return ((x ** 2) + (9 * y ** 2) / 4 + (z ** 2) - 1) ** 3 - (x ** 2) * (z ** 3) - ((9 * (y ** 2) * (z ** 3)) / 200)


def draw(figure_func, bbox=(-1.3, 1.3)):
    xmin, xmax, ymin, ymax, zmin, zmax = bbox * 3
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(projection='3d')
    a = np.linspace(xmin, xmax, 80)
    b = np.linspace(ymin, ymax, 80)
    a1, a2 = np.meshgrid(a, a)
    for x in b:
        y, z = a1, a2
        X = figure_func(x, y, z)
        cest = ax.contour(X + x, y, z, [x], zdir='x', colors='red')
    for y in b:
        x, z = a1, a2
        Y = figure_func(x, y, z)
        cest = ax.contour(x, Y + y, z, [y], zdir='y', colors='red')

    for z in b:
        x, y = a1, a2
        Z = figure_func(x, y, z)
        cest = ax.contour(x, y, Z + z, [z], zdir='z', colors='red')

    ax.set_xlim3d(xmin, xmax)
    ax.set_ylim3d(ymin, ymax)
    ax.set_zlim3d(zmin, zmax)
    plt.show()


# напишите код здесь
draw(figure_func=func)
```

$3$. Построить линейный график ["Белого шума"](https://ru.wikipedia.org/wiki/%D0%91%D0%B5%D0%BB%D1%8B%D0%B9_%D1%88%D1%83%D0%BC), представленного на изображении размером 1000x1000 пикселей. Генерация шума должна осуществляться на основе выдачи случайных значений по координатам пикселей исходного изображения. При реализации функции шума `noise()` запрещается использование модуля `random`. Для построения графика задать параметр `figsize=(18,6)`.

```python
from math import sin, floor


def noise(x, y):
    return fract(sin(x * 12.9898 + y * 78.233) * 43758.5453123)


def fract(x):
    return x - floor(x)


xs = []
ys = []
for x in range(0, 1000):
    y = x
    xs.append(x)
    ys.append(noise(x, y))

plt.figure(figsize=(18, 10))
plt.plot(xs, ys)
```
